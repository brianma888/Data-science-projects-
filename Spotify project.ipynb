```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
%matplotlib inline
import warnings 
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv(r'top50.csv',encoding='latin-1')
```

```python
df
```

```python
df.describe()
```

```python
df.columns=['Rank', 'Track.Name', 'Artist.Name', 'Genre', 'Beats.Per.Minute',
       'Energy', 'Danceability', 'Loudness..dB..', 'Liveness', 'Valence.',
       'Length.', 'Acousticness..', 'Speechiness.', 'Popularity']
```

```python
df
```

```python
category_genre=pd.DataFrame(df['Genre'].value_counts())
```

```python
category_genre.head()
```

```python
plt.figure(figsize=(15,10))
sns.barplot(x='Genre',y=category_genre.index,data=category_genre)
plt.title('Top 50 songs on Spotify by Genre')
plt.xlabel('Number of songs')
plt.ylabel('Genre')

#It can be seen that dance pop has the highest number of songs, and big room has the least number of top 50 songs 
```

```python
groupby_genre=df.groupby(by='Genre').mean()
beats=pd.DataFrame(groupby_genre['Beats.Per.Minute'])
beats.reset_index(inplace=True)
beats.columns=['genre','beats per minute']
beats

sns.set(font_scale=1)
sns.set_style('darkgrid')
plt.figure(figsize=(40,30))
plt.tight_layout()
sns.scatterplot(x='genre',y='beats per minute',data=beats,s=200)
```

```python
top10_artist=pd.DataFrame(df['Artist.Name'].value_counts().head())
top10_artist.reset_index(inplace=True)
top10_artist.columns=['name of artist','number of songs']
top10_artist

plt.figure(figsize=(15,10))
sns.barplot(x='name of artist',y='number of songs',data=top10_artist)
```
