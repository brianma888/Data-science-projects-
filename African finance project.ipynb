```python
Algeria_stats=df[df['country']=='Algeria']
Algeria_stats.head()
```

```python
Algeria_stats.tail(1)
```

```python
sns.set_style('darkgrid')
plt.figure(figsize=(5,5))
sns.lineplot(data=Algeria_stats,x='year',y='exch_usd',color='red')
plt.title('Algeria dollars usd exchange rate trend from 1870 to 2014 ')
```

```python
Angolo_stats=df[df['country']=='Angola']
plt.figure(figsize=(5,5))
sns.lineplot(data=Angolo_stats,x='year',y='exch_usd')
```

```python
plt.figure(figsize=(15,10))
sns.lineplot(data=Algeria_stats,x='year',y='exch_usd',color='red')
sns.lineplot(data=Angolo_stats,x='year',y='exch_usd')
sns.lineplot(data=df[df['country']=='Central African Republic'],x='year',y='exch_usd')
sns.lineplot(data=df[df['country']=='Ivory Coast'],x='year',y='exch_usd')
sns.lineplot(data=df[df['country']=='Egypt'],x='year',y='exch_usd')
sns.lineplot(data=df[df['country']=='Kenya'],x='year',y='exch_usd')
sns.lineplot(data=df[df['country']=='Mauritius'],x='year',y='exch_usd')
sns.lineplot(data=df[df['country']=='Morocco'],x='year',y='exch_usd')
sns.lineplot(data=df[df['country']=='Nigeria'],x='year',y='exch_usd')
sns.lineplot(data=df[df['country']=='South Africa'],x='year',y='exch_usd')
sns.lineplot(data=df[df['country']=='Tunisia'],x='year',y='exch_usd')
sns.lineplot(data=df[df['country']=='Zambia'],x='year',y='exch_usd')
sns.lineplot(data=df[df['country']=='Zimbabwe'],x='year',y='exch_usd')
leg=plt.legend(['Algeria','Angolo','Central African Republic','Ivory Coast','Egypt','Kenya','Mauritus','Morocco','Nigeria','South Africa','Tunisia','Zambia','Zimbabwe'])
             
```
