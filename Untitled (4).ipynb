```python
states_number=df.groupby(by='state').sum().sort_values(by='number',ascending=False)
```

```python
states_number
```

```python
states_number.reset_index(inplace=True)
```

```python
states_number
```

```python
import seaborn as sns 
```

```python
sns.set_style('darkgrid')
```

```python
fig=sns.scatterplot(x=states_number['state'],y=states_number['number'])
plt.xticks(rotation=90)
```

```python
fig
plt.show()
```

```python
states_number
```

```python

sns.barplot(data=states_number,x='state',y='number',color='blue',hue='year')
plt.xticks(rotation=90)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
```
